# Plotting Bar Charts , Pie Charts and Histograms 

#### Bar Chart : We use bar chart to compare the two or more categroies side by side  --- Date Analysis
#### Pie Chart : We use Pie Chart to see which part is taking most of the percentage , proportion  Illustration --- Whole Representation
#### HistoGram : We use it to find how much data is spread across ranges , Numerical Distributions  ---Date Insights 


# Bar Chart : We use bar chart to compare the two or more categroies side by side  --- Date Analysis


```python
# Bar Chart 
# plt.bar(x,height,color="color name",width = value , label = "") x will be the category , and height will be the value of that category to show the ehight 
# The bar Chart Can be Horizonal and Vertical too 
product = ['A','B','C','D']
sales = [1000,1400,3420,30]

plt.bar(product,sales,color="red",label = "Categroy")
# plt.barh(product,sales,color="red",label = "Categroy") we use this barh to make teh horizontal bars 
plt.xlabel("Product")
plt.ylabel("Sales")
plt.title("Bar Chart Representation")
plt.legend()
plt.show()

```

# Pie Chart : We use Pie Chart to see which part is taking most of the percentage , proportion  Illustration --- Whole Representation
 

```python
# plt.pie(values,labels = labels_list , colors  = color_list , autopct='%1.1f%%' ) this autopct = %1.1f%% means that .1f means we will only take the value 1 after the decimal , and %% means we have to show the Percentage(%) sign 

regions = ['North','South','East','West']
revenue = [3000,2000,1500,1000]

plt.pie(revenue , labels = regions , autopct = '%1.1f%%',colors=['gold','skyblue','lightgreen','pink'])
plt.title("Pie Chart Representation")
plt.show()
```

# HistoGram : We use it to find how much data is spread across ranges , Numerical Distributions  ---Date Insights

```python
# plt.hist(data,bins=number_of_bins,color='colorname',edgecolor='black')
# here bins means no. of distributions you want to see. edge color is the border color of each bar , data is the value data
scores = [43,24,34,5,6,7,56,32,13,4,7,87,8,54,5,54,89,24,46,64,5,55,43,45]
plt.hist(scores,bins=8,color="purple",edgecolor="black")
plt.xlabel("Score Range")
plt.ylabel("No. Of Student")
plt.title("Histogram Representations ")
plt.show()
# height of the bar = no. of students in that range 
```

# Scatter Plot 

#### used to find the correlation between 2 variable for eg: AD Budget vs sales , temperature vs ice sales
markers possible values : <br>
"o" for circle <br>
"s" for square<br>
"^" for triangle <br>
"D" for Diamond<br>
"x" for cross<br>
"+" for plus sign<br>

```python
# plt.scatter(x,y,color="Color Name",marker="marker style",label = "label name ")

hours = [1,2,3,4,5,6,7,8]
exam_score = [50,55,60,65,70,75,80,85]

plt.scatter(hours, exam_score,color="red",label = "Student Data" , marker="^" )
plt.xlabel("Hours")
plt.ylabel("Exam Scores")
plt.title("Scatter Plot Representation")
plt.grid()
plt.legend()
plt.show()
```

# Using 2 or more dataset to represent on the graph

```python
rollNo = [1,2,3,4,5,6,7,8]
ScienceMarks = [55,66,85,74,51,58,81,12]
MathsMarks = [45,34,78,34,23,89,11,9]

plt.scatter(rollNo,ScienceMarks,color="red",marker="o" , label="Science Marks")
plt.scatter(rollNo,MathsMarks,color="Blue",marker="s",label="Maths Marks")

plt.xlabel("Roll Number")
plt.ylabel("Subject Marks")
plt.title("Student Marks Representation")
plt.legend(loc="upper right")
plt.xlim(1,9)
plt.ylim(0,100)
plt.grid()
plt.show()
```

# Advance Matplotlib  
### SubPlots And Layout Adjustments

```python
# Subplots 
# plt.subplot(nrows,ncols,index) here is 1 based indexing 

x = [1,5,9,2]
y = [10,25,35,89]
plt.subplot(1,2,1) # here it means that 1st row , 2nd column , 1st subplot
plt.plot(x,y)
plt.title("Line Chart ")

plt.subplot(1,2,2) # here it means that 1st row , 2nd column , 2nd subplot
plt.bar(x,y)
plt.title("Bar Chart")

# plt.tight_layout()

plt.show()
```

# Recommanded Way to do in the interview.

```python
# fig, ax=plt.subplots(nrow,ncol,figsize=(width,height)) here fig, is important as it is used to do the figure.
# fig.suptitle("Title Here") used to give the title to the whole figure
# ax[index].graphType(x,y,color="color name ")
fig, ax=plt.subplots(1,2,figsize=(10,5))
x = [1,5,9,2]
y = [10,25,35,89]

ax[0].plot(x,y,color="red",label="Line Chart ") # used to plot the subgraph at the index 0 
ax[0].set_title("Line Chart ") # used to set the title of the graph



ax[1].bar(x,y,color="green",label="bar Chart") # used to plot the subgraph at the index 1
ax[1].set_title("Line Chart ") # used to set the title of the bar chart


fig.suptitle("Two or More Charts Show Up ") # used for give the super title of whole figure.
plt.legend()


plt.tight_layout() # this is used to enlarge the graph to take up the good amount of space. 
# you can say for auto adjustment overlap fix , and also ways used just before the plt.show()


plt.show()
```

# Saving The Figures using -- savefig()

```python
# savefig('filename.extension',dpi=value,bbox_inches='tight')  here it means that dpi means image resolution , and filename is name of the file you want to name it, and bbox_iches means cropping the white spaces 

x=[1,2,3,4]
y=[12,23,45,23]
plt.plot(x,y,color="red",marker='o',label="For Saving Image")
plt.title("Image for saving")
plt.xlabel("X")
plt.ylabel("Y")


plt.savefig('LineChart.png',dpi=300,bbox_inches="tight") # used to save the image of the plot in whatever extention you have given.
# To Save the File in a existing Folder 
# Here we have to make sure that we have to only write the folder name which already exist.
plt.savefig('FolderName/LineChart.png',dpi=300,bbox_inches="tight") 
```
