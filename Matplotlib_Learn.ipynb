# Importing Matplotlib - pyplot as plt 

```python
import matplotlib.pyplot as plt
```

# Making First Plot and Showing it 

```python
x = [1,2,3,4]
y = [10,20,15,25]

plt.plot(x,y)   # .plot(x,y) it takes 2 parameters which are horixontal and vertical values of the graph 

plt.show()  # .show() it is used to show the output of the graph.
```

```python
x = ["Mon","Tue","Wed","Thru","Fri","Sat"]
y = [100,36,8,54,3,4]
plt.plot(x,y)
plt.title("Backery Sales of Week.")
plt.xlabel("Day of the week ")
plt.ylabel("Sales per day ")
plt.show()
```

# Pyplot Function : MATPLOTLIB PLOTTING FUNCTIONS

```python
# plt.plot(x,y,color="Color_Name",linestyle="line_style",linewidth=value , marker='marker_symbol',label='label_name')
# x,y are mentadory parameter and rest are optional parameters. 
# plt.xlabel("Text")
# plt.ylabel("Text")
# plt.title("Name or text ")
# plt.legend(loc = "position" ,fontsize=value)  these loc and fontsize parameters are optional , and in loc we can use "upper left" "upper right" "lower left" "lower right"
# plt.grid(color="color name " , linestyle="" , linewidth=value)
# plt.xlim(start,end) here end will be included these functions will limit the values which are used to represent the graph values 
# plt.ylim(start,end) end will be included these functions will limit the values which are used to represent the graph values 
# plt.xticks(LIST OF X AXIS VALUES ,LIST OF X AXIS NAME CHANGING VALUES) this function will help you name the values used to represent the graph on the x axis , by giving them meaning full names 
# plt.yticks() 
# plt.show() this funciton helps you to show or display the graph which oyu have created or modified to see the changees or graph 

month = [1,2,3,4]
sales = [100,200,4523,345]

plt.plot(month,sales,color="blue",linestyle="-",linewidth=2,marker='o',label="2025 Salles ")
plt.title("plt.plot()")
plt.xlabel("month")
plt.ylabel("sales")
plt.xlim(1,4)
plt.ylim(0,5000)
plt.grid(color="red" , linestyle=":") # there  linestyle=":" means dotted line 
plt.xticks([1,2,3,4],['M1','M2','M3','M4'])
# plt.yticks([100,200,4523,345],['P1','P2','P3','P4']) commented becuase the doesnt look good due to this functions
plt.legend()
plt.show()
```
