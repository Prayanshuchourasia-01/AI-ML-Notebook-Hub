# Importing Matplotlib - pyplot as plt 

```python
import matplotlib.pyplot as plt
```

# Making First Plot and Showing it 

```python
x = [1,2,3,4]
y = [10,20,15,25]

plt.plot(x,y)   # .plot(x,y) it takes 2 parameters which are horixontal and vertical values of the graph 

plt.show()  # .show() it is used to show the output of the graph.
```

```python
x = ["Mon","Tue","Wed","Thru","Fri","Sat"]
y = [100,36,8,54,3,4]
plt.plot(x,y)
plt.title("Backery Sales of Week.")
plt.xlabel("Day of the week ")
plt.ylabel("Sales per day ")
plt.show()
```
